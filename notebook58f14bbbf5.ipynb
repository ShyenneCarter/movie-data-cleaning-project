```python
# Convert Release_Date to datetime
df["Release_Date"] = pd.to_datetime(df["Release_Date"], errors="coerce")

# Convert Vote_Count and Vote_Average to numeric
df["Vote_Count"] = pd.to_numeric(df["Vote_Count"], errors="coerce")
df["Vote_Average"] = pd.to_numeric(df["Vote_Average"], errors="coerce")
df["Title"] = df["Title"].fillna("Unknown")
df["Overview"] = df["Overview"].fillna("Unknown")
df["Original_Language"] = df["Original_Language"].fillna("Unknown")
df["Genre"] = df["Genre"].fillna("Unknown")
df["Poster_Url"] = df["Poster_Url"].fillna("Unknown")

df["Popularity"] = df["Popularity"].fillna(0)
df["Vote_Count"] = df["Vote_Count"].fillna(0)
df["Vote_Average"] = df["Vote_Average"].fillna(0)

df["Title"] = df["Title"].str.strip().str.title()
df["Overview"] = df["Overview"].str.strip()
df["Original_Language"] = df["Original_Language"].str.strip().str.lower()
df["Genre"] = df["Genre"].str.strip().str.title()

df = df[(df["Vote_Average"] >= 0) & (df["Vote_Average"] <= 10)]
df = df[df["Vote_Count"] >= 0]
df = df[df["Popularity"] >= 0]

df.info()
df.isnull().sum()
df.duplicated().sum()

```

```python
df.to_csv("cleaned_movies.csv", index=False)
df.info()
```
